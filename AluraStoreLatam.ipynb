# 2. Ventas por categoría

```python
import pandas as pd

# Cargar los conjuntos de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [(tienda, "Tienda 1"), (tienda2, "Tienda 2"), (tienda3, "Tienda 3"), (tienda4, "Tienda 4")]

def contar_ventas_por_categoria(df):
    categorias = df['Categoría del Producto'].tolist()  # Convertir columna "Categoría del Producto" a lista
    conteo = []
    categorias_unicas = list(set(categorias))  # Obtener categorías únicas

    for cat in categorias_unicas:
        if pd.notna(cat):  # Ignorar valores nulos
            conteo.append([cat, categorias.count(cat)])

    conteo.sort(key=lambda x: x[1], reverse=True)
    return conteo

for df, nombre_tienda in tiendas:
    print(f"\n{nombre_tienda}:")
    ventas_categoria = contar_ventas_por_categoria(df)

    print(" Categoría         Cantidad Vendida")
    print("-" * 35)
    for cat, cantidad in ventas_categoria:
        print(f" {cat:<15} {cantidad:>5}")

    categoria_mas_popular = ventas_categoria[0]
    print(f"\nCategoría más popular en {nombre_tienda}: {categoria_mas_popular[0]} "
          f"con {categoria_mas_popular[1]} ventas.\n")
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

# cargar los datos de las tienndas
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# lista con las tienndas
tiendas = [(tienda, "Tiennda 1"), (tienda2, "Tiennda 2"), (tienda3, "Tiennda 3"), (tienda4, "Tiennda 4")]

# prosesar tienda 1
print("\nTiennda 1:")
productos = tienda['Producto'].tolist()  # sacar lista de productos
lista_productos = []
productos_diferentes = list(set(productos))  # productos sin repetir
for prod in productos_diferentes:
    if prod == prod:  # no contar nulos
        cantidad = productos.count(prod)
        lista_productos.append([prod, cantidad])
# ordenar de mayor a menor
lista_productos.sort(key=lambda x: x[1], reverse=True)
# mostrar mas vendido y menos vendido
if lista_productos:
    print(f"Producto mas vendido: {lista_productos[0][0]} con {lista_productos[0][1]} venttas")
    print(f"Producto menos vendido: {lista_productos[-1][0]} con {lista_productos[-1][1]} venttas")
    print("\nLista de productos:")
    print("Producto         Cantidad")
    print("-----------------------")
    for prod, cant in lista_productos:
        print(f"{prod:<15} {cant:>5}")

# prosesar tienda 2
print("\nTiennda 2:")
productos = tienda2['Producto'].tolist()
lista_productos = []
productos_diferentes = list(set(productos))
for prod in productos_diferentes:
    if prod == prod:
        cantidad = productos.count(prod)
        lista_productos.append([prod, cantidad])
lista_productos.sort(key=lambda x: x[1], reverse=True)
if lista_productos:
    print(f"Producto mas vendido: {lista_productos[0][0]} con {lista_productos[0][1]} venttas")
    print(f"Producto menos vendido: {lista_productos[-1][0]} con {lista_productos[-1][1]} venttas")
    print("\nLista de productos:")
    print("Producto         Cantidad")
    print("-----------------------")
    for prod, cant in lista_productos:
        print(f"{prod:<15} {cant:>5}")

# prosesar tienda 3
print("\nTiennda 3:")
productos = tienda3['Producto'].tolist()
lista_productos = []
productos_diferentes = list(set(productos))
for prod in productos_diferentes:
    if prod == prod:
        cantidad = productos.count(prod)
        lista_productos.append([prod, cantidad])
lista_productos.sort(key=lambda x: x[1], reverse=True)
if lista_productos:
    print(f"Producto mas vendido: {lista_productos[0][0]} con {lista_productos[0][1]} venttas")
    print(f"Producto menos vendido: {lista_productos[-1][0]} con {lista_productos[-1][1]} venttas")
    print("\nLista de productos:")
    print("Producto         Cantidad")
    print("-----------------------")
    for prod, cant in lista_productos:
        print(f"{prod:<15} {cant:>5}")

# prosesar tienda 4
print("\nTiennda 4:")
productos = tienda4['Producto'].tolist()
lista_productos = []
productos_diferentes = list(set(productos))
for prod in productos_diferentes:
    if prod == prod:
        cantidad = productos.count(prod)
        lista_productos.append([prod, cantidad])
lista_productos.sort(key=lambda x: x[1], reverse=True)
if lista_productos:
    print(f"Producto mas vendido: {lista_productos[0][0]} con {lista_productos[0][1]} venttas")
    print(f"Producto menos vendido: {lista_productos[-1][0]} con {lista_productos[-1][1]} venttas")
    print("\nLista de productos:")
    print("Producto         Cantidad")
    print("-----------------------")
    for prod, cant in lista_productos:
        print(f"{prod:<15} {cant:>5}")
```
